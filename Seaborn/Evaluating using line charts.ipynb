```python
import pandas as pd
import seaborn as sb
```

```python
a=pd.read_excel("/content/Sample_Sales Records_xls.xls")
a.head()
```

```python
sb.relplot("Total Revenue","Total Cost",data=a,kind="line")
```

```python
sb.relplot("Order Date","Total Revenue",data=a,kind="line")
```

```python
# For standard deviation
sb.relplot("Order Date","Total Revenue",data=a,kind="line",ci="sd",color="m")
```

```python
# Canceling the estimator
sb.relplot("Order Date","Total Revenue",data=a,kind="line",estimator=None)
```

```python
sb.relplot("Order Date","Total Revenue",data=a,kind="line",estimator=sum,hue="Sales Channel")
```

```python
sb.relplot("Order Date","Total Revenue",data=a,kind="line" ,style="Sales Channel",hue="Order Priority")
```

```python
sb.relplot("Order Date","Total Revenue",data=a,kind="line" ,col="Sales Channel",hue="Order Priority")
```

```python
sb.relplot("Order Date","Total Revenue",data=a,kind="line" ,hue="Sales Channel",col="Item Type",col_wrap=4)
```
