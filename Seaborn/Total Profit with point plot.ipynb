```python
# Point Plot
#The point plot in seaborn means a scatter plot depicting point estimations for categories 
import pandas as pd
import seaborn as sb
```

```python
a= pd.read_excel("/content/Sample_Sales Records_xls.xls")
a.head(7)
```

```python
# catplot
sb.catplot(x="Total Revenue",y="Order Priority",data=a,kind="point")
```

```python
sb.catplot(x="Total Revenue",y="Item Type",data=a,hue="Order Priority",kind="point")
```

```python
sb.catplot(col="Item Type",hue="Order Priority",x="Sales Channel",data=a,y="Total Profit",kind="point",col_wrap=4)
```
