```python
import matplotlib.pyplot as pt
```

```python
# Area plot / Stackplot
# Area charts are commonly used to showcase data that depicts a time-series relationship.
days=[1,2,3,4,5]
Coding=[8,5,7,8,13]
Development=[7,8,6,11,7]
WorkOut=[2,3,4,3,2]
Academics=[7,8,7,2,2]
```

```python
pt.plot([],[],color='c',label='Coding',linewidth='5')
pt.plot([],[],color='g',label='Development',linewidth='5')
pt.plot([],[],color='b',label='WorkOut',linewidth='5')
pt.plot([],[],color='r',label='Academics',linewidth='5')

pt.stackplot(days,Coding,Development,WorkOut,Academics,colors=['c','g','b','r'])
pt.legend()
```

```python
# Piechart
# The size of each slice of the pie is proportional to the number of data values in the corresponding class.
slices=[13,7,2,2]                       # Relevant sizes for each pie
```

```python
activities=['Coding','Development','WorkOut','Academics']
col=['c','g','b','r']
```

```python
pt.pie(slices,labels=activities,colors=col,startangle=90,shadow=True,   # 90 means first division is vertical line
       explode=(0,0,0,0.1),autopct='%1.1f%%')                           # explode to pull out slice a bit
                                                                        # actopct overlays percentage over piechart
```
