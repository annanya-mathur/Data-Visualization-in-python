```python
import matplotlib.pyplot as pt
```

```python
from mpl_toolkits.mplot3d import axes3d
```

```python
pt.ion()
```

```python
y=[4,3,1,1,3,4]
```

```python
x=[2,1,1,3,3,2]
```

```python
y2=[8,7,6,5,6,7,8]
```

```python
x2=[6,5,5,6,7,7,6]
```

```python
pt.plot(x,y)   # polygon in 2D
pt.plot(x2,y2)
pt.xlabel('x')
pt.ylabel('y')
pt.title('My Polygon')
pt.legend(['f(x) pentagon','f(x2) hexagon'])
```

```python
f=pt.figure()  # 3-D plot of polygons
a=f.add_subplot(111,projection='3d')  # 111 for height, breadth, length
a.plot(x,y)
a.plot(x2,y2)
a.legend(['f(x) pentagon in 3D','f(x2) hexagon in 3D'])
```
