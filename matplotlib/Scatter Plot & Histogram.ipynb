```python
# Histogram
# Histogram have quantitative values but bar graph has categorical data
import matplotlib.pyplot as pt
ages=[21,20,29,15,40,49,45,56,67,46,43,42,45,49,40,25,30,39,34,44,70,71,70,75,43,43,41,40,56,43,44,47,65,64,66,66,47,48]
points=[10,20,30,40,50,60,70]
pt.hist(ages,points,histtype='stepfilled',rwidth=0.9)                           # blue- b            purple- m
pt.xlabel('Age group',color='m')                                                # red- r             black- k
pt.ylabel('Percentage of contributing to development ',color='m')               # green- g           white- w
pt.title('Histogram for project devepolment')                                   # mustard- y         sky blue- c
```

```python
# Scatter Plot
# It is used to find relation b/t two or more variables
x=[1,2,3,4,5,6,7,8,9,0]
y=[5,4,3,2,1,5,1,7,3,9]
pt.scatter(x,y,color='b')
pt.legend('Scatter Points')
pt.xlabel('x-axis',color='m')
pt.ylabel('y-axis',color='m')
```

```python
# making downwards pointing triangle inplace of points                          # point -'.'
pt.plot(x,y,'vr')                                                               # plus -'+'
```

```python
# for joining scatter points through line
pt.plot(x,y,'-',color='m')
```
