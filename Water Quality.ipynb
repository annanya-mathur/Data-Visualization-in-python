# **Analysing** **WATER Quality**
---

1.   By Checking which State has highest PH values
2.   By identifying the year which suffered more water pollution
3.   Finding Corelation between different parameters







```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as pt
import seaborn as sb
import datetime as dt
```

```python
df=pd.read_excel('/content/water.xlsx')
```

```python
df.head(5)
```

```python
df.tail()
```

```python
df=df.sort_values(by="PH",ascending=False)
df.head()
```

```python
sb.catplot(y='STATE',x='PH',data=df,kind='point')
pt.grid(True ,color='k')
```

**Hence, Almost all states have ph > 7 i.e. water bodies are polluted** 

danger.jpg






```python
sb.barplot(x="year",y="PH",data=df)
sb.set(rc={'figure.figsize':(10,8)})
pt.grid(True,color='k')
```

**Hence in year 2010 water bodies were found polluted in many states.**

```python
sb.catplot(y='STATE',x='B.O.D. (mg/l)',data=df,kind='point')
pt.grid(True ,color='k')
```

```python
sb.catplot(y='STATE',x='D.O. (mg/l)',data=df,kind='point')
pt.grid(True ,color='k')
```

**Dissolved Oxygen Value less than 4 (mg/l) leads to adverse environment for aquatic life as in case of haryana.**

```python
pt.figure(figsize=(12,8))
sb.heatmap((df).corr(),annot=True,fmt='.0%')
```

**Correlation Chart**
