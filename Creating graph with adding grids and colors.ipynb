```python
import matplotlib.pyplot as pt
```

```python
import matplotlib.style as s                   # adding grid and colors to graph
```

```python
s.use('ggplot')                                # data visualization package
```

```python
x1=[3,8,5]
y1=[4,6,1]
x2=[5,8,9]
y2=[10,4,8]
```

```python
pt.plot(x1,y1,'g',linewidth=6)
pt.plot(x2,y2,'r',linewidth=6)
pt.legend(['data1','data2'])
pt.grid(True,color='k')
pt.xlabel ('No. of people',color='k')
pt.ylabel('Sales',color='k')
pt.title('Annual Sales',color='k')
```
